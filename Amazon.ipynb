```python
import pandas as pd
```

```python
df=pd.read_csv('amzone.csv',usecols=['Id','Score','Text'],on_bad_lines='skip')
df=df.sample(500,random_state=42)
```

```python
df
```

```python
import re
import nltk
nltk.download('all')
```

```python
from nltk.corpus import stopwords #a ,and ,the
from nltk.tokenize import word_tokenize
stopwords_set=set(stopwords.words('english'))
```

```python
def Clean_Txt(text):# I LOVE THe Food <3
  text=text.lower()#  i love the food <3
  text=re.sub(r'[^a-z\s]',' ',text)#  i love the food
  tokens=word_tokenize(text) #["i","love","the","food"]
  tokens=[i for i in tokens if i not in stopwords_set] #["love","food"]
  return ' '.join(tokens)# love food
```

```python
df["cleaned_text"]=df['Text'].apply(Clean_Txt)
```

```python
from textblob import TextBlob
def sentiment(text):
  p=TextBlob(text).sentiment.polarity
  if p>0:
    return 'Positive'
  elif p<0:
    return 'Negative'
  else:
    return "Neutral"
df["sentiment"]=df['cleaned_text'].apply(sentiment)
```
